## Arboles de Decision

```python
import urllib.request
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

import numpy as np
```

```python
# Cargamos los datos y los dividimos en conjuntos para entrenar y testear.
data = df
y = data["Fallecido"]
X = data.drop(["Fallecido"], axis=1)
(X_train, X_test,
 y_train, y_test) = train_test_split(X,
                                     y,
                                     test_size=0.33,
                                     random_state=42)
```

```python
# Primero creamos un arbol de decisión para ver la precisión y poder compararla
tree = DecisionTreeClassifier(max_depth=2, random_state=11) # Max depth acorta el arbol si no salia muy grande.
tree.fit(X_train, y_train)
print(f"La precisión del Arbol de decisión es de: {tree.score(X_test, y_test)}")
```

#### Osea el modelo es capas de predecir correctamente el 76% de las muestras del conjunto de prueba. Es un rendimiento razonable para seguir analizando.

```python
tree.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree # Importamos para graficar el arbol de decisión.
import matplotlib.pyplot as plt # Para manejar el tamaño del grafico del arbol.
```

### A) Visualizando el modelo

```python
plot_tree(decision_tree=tree, feature_names=X_train.columns, filled=True, fontsize=12); 
plt.figure(figsize=(10,5))

# Feature=nombre de columnas, 
# Filled=Para colorear el arbol, 
# El punto y coma al final es para que no salga todo el texto del arbol de decision y solo el grafico
```

#### La condicion principal es "Tiempo", si el valor tiempo es menor a 67,5 se sigue por la rama derecha en este caso, por el contrario sera por la izquierda. Luego si "Creatinina Suero" es menor o igual a 0.194 se sigue por la izquierda, llegando al final de la rama en este caso. Tenemos "value" "[133,9] y "samples=142" tenemos que dividir 133/142 para saber que la probabilidad de que salga en esta rama es de 93.66%
#### Cuanto más cercano a 0 sea el valor de Gini, mayor será la pureza en términos de clasificación de las muestras en esa hoja. Un valor más alto indica una mezcla de diferentes clases en las muestras.
